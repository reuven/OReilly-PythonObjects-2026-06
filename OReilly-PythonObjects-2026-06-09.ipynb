# What are objects?

Software is hard to write. And it's even harder to maintain.

In the 1970s, people already found this to be the case, and they started to look for ways to make software easier to write, and (especially) easier to maintain.

Alan Kay, at Xerox PARC, invented a programming language called Smalltalk. It was the first language to use "objects," which were a new paradigm meant to make it easier to write and maintain software.

Software was seen as very complex. Kay and his colleagues thought about what complex things existed in the world that they could perhaps make software more similar to. They decided to look at biological systems:

In biology, we have cells. Each cell has a particular type. Each type of cell knows how to send and receive different types of messages. If you know what kind of cell you're looking at, you know what its interface is to the world.

What if:

- We design our code using "objects," which are basically cells.
- Each object has as type, or a "class," which defines its behavior.
- Instead of sending/receiving messages, we'll talk about "invoking methods."

This was the paradigm shift that Kay promoted with Smalltalk.

A very large number of programming languages today now use this paradigm of objects, and they all owe a debt of gratitude to Smalltalk, even if they don't say so.

Python is one such language -- and some of the ideas and terminology in Python are actually taken from Smalltalk.

# Who cares? Or: So what?

We say that everything is an object in Python. What does this mean?

It means several things:

- The language is consistent. Everything follows the same rules, whether it's built-in stuff that comes with the language, or things that we add to the language as extensions.
- We can extend the language by writing new classes (i.e., new data types) that follow these rules

Once you understand how Python objects work, you'll see so many places in the core language that make more sense to you. 

Why are objects hard for people to learn?

- The vocabulary, which is different from the rest of programming
- It's treated as a religious discipline. It is a packaging, or logistical, technique, to make it easier to think about our software as we write, read, and maintain it.

# Object jargon/vocabulary

- `class` and `type`. Every value in Python is an object, and every object has a *type*. This means that every object's behavior is defined by certain rules, rules that are defined in the type. If I have a string, I know what a string does, because of the `str` *class*. In verbal Python, you can talk about "class" and "type" interchangeably. In Python code, they each have a specific role; you'll define new 
data types with the `class` keyword.
- You can think of a class as a factory for creating new objects, or new values. `str` is a factory for string values, and `dict` is a factory for dictionary values.
- When we talk about one particular string, we're talking about an *instance* of `str`. The class (type) is `str`, and the instance is one particular string value. In other words, the class (factory) produces many instances, each belonging to that class.
- Python has millions of objects running around at any given time. Every single one of them is an instance of some class. The class defines the behavior.

If I want to create a new value of type `X`, where `X` is the class/type, then I invoke `X()`. This gives me a new value of `X` back, aka a new instance of `X`.
    - I can create a new integer with `int()`
    - I can create a new string with `str()`
    - Many people say that `int` converts a string to an integer. But no! It creates a new instance of `int`, based on its argument (which can be a string).

- `instance` -- the word means, "a value whose class is ___," where ___ is a class.
- `object` -- this word is used in *far* too many different contexts. Every value in Python is an object. Every class is descended (through inheritance, which we'll discuss), from the top of the hierarchy, which is known as `object`. 

# What does the class matter?

# More slang!

If I say "dunder init," that's Python slang for `__init__` -- double underscore before and after the word "init". Dunder methods (like `__init__` are crucial in the Python ecosystem. If you define a dunder method, then either Python will run it automatically, or you're changing the default behavior. You are typically not going to run these methods yourself, directly.

```python
# let's define a more complex Person class

class Person:
    def __init__(self, first_name, last_name, shoe_size):
        self.first_name = first_name
        self.last_name = last_name
        self.shoe_size = shoe_size
```

```python
p = Person('Reuven', 'Lerner', 46)
```

```python
p.shoe_size
```

```python
p.first_name
```

# Attributes

What the heck is an attribute?

In Python, we know that we can store values in a *variable*. But it turns out that Python has *two* storage mechanisms. Variables are just one. The other is *attributes*.

Every single object in Python has attributes. They are a private dictionary on every object value. It doesn't look like a dict, but every attribute has a name and a value.

The names are normally written after a variable name, separated by a `.`. The values can be anything at all, no restrictions.

If I say

    a.b

this means, to Python, that I want to go to `a`, grab the value associated with that variable, and then turn the value and ask, "Do you have an attribute `b`?" If it does, then I want the value.

You can always tell an attribute, beacuse it comes after a `.`.

In other langauges, you'll hear them talk about "instance varibles," "class variables," "fields," etc. And in Python, we have one way to store things on our objects, namely *attributes*.

When we say `self.name = name`, we're really saying:

- Get the value of the `name` parameter (i.e., local variable)
- Assign its value to a new attribute on `self` (the new instance of `Person`).
- The new attribute's name will be `name` and its value will be whatever the `name` parameter refers to.
- Once we've done that, and `name` is an attribute on `self`, then we'll be able to retrieve it later on, too.
- `__init__` is invoked automatically every time a new object is created. Its job is to add attributes (and values) to a newly created object.


```python
class Person:
    def __init__(self, first_name, last_name, shoe_size):
        self.first_name = first_name   # take the value from the first_name variable, and use it to create a first_name attribute on self
        self.last_name = last_name      # take the value from the last_name variable, and use it to create a last_name attribute on self
        self.shoe_size = shoe_size     # take the value from the shoe_size variable, and use it to create a shoe_size attribute on self
```

```python
p = Person('Reuven', 'Lerner', 46)
```

```python
p.first_name
```

```python
p.last_name
```

```python
p.shoe_size
```

You want to assign *all* of the attributes that you'll ever want on your object in `__init__`, even if that means assigning `None` or a similar value to one or more attributes. This just makes your code easier to read and maintain.

# What is `self`?

- `self`, in a method, refers to the instance itself. It's your way for the method to retrieve and set attributes on the instance.
- Many other languages use the name `this`, rather than `self`.
- In other languages, `this` (or even `self`) is a reserved word, and special. In Python, it's a convention -- you shouldn't use a different name, but you aren't technically forced to use `self`.
- If we want to assign a new attribute to our instance from within a method, we assign to `self`
- If we want to retrieve an attribute from our instance from within a method, we retrieve from `self`
- `self` is just a parameter, and always the first one. Why not make it a reserved word? To avoid confusing things even more, with scoping and priorities.
- Notice that we invoke `Person` with three arguments, but `__init__` requires four. Python automatically inserts the instance as the first argument, which is then assigned to `self`.

# What about other methods?

So far, we have one method, `__init__`, which is invoked by Python automatically when we have a new instance.

It might be nice to have other methods - for example, the `summary` method that we had before. 

Or maybe we might want methods to return the person's name or other info.

In many (most?) programming languages that use objects, it's *standard* to create two methods for every attribute: You have a "getter" method (sometimes known as an "accessor") that returns it, and a "setter" method (sometimes known as a "mutator") that modifies it.

```python
class Person:
    def __init__(self, name):
        self.name = name

    def get_name(self):
        return self.name   # you cannot just return "name" -- it has to be "self.name"

    def set_name(self, new_name):    # get a new name
        self.name = new_name         # set the new name to the existing attribute
```

```python
p = Person('Reuven')  # create the instance
```

```python
p.get_name()   # get the name
```

```python
p.set_name('newReuven')
p.get_name()
```

# Getters and setters

We almost never write getters and setters in Python.

Why not? Because we don't need to!

- We can retrieve the value of an attribute just by accessing it -- `p.name`
- We can set the attribute by setting it -- `p.name = 'something else'`

```python
class Person:
    def __init__(self, name):
        self.name = name

p = Person('Reuven')  # create the instance
p.name
```

```python
p.name = 'newReuven'
p.name
```

What about all sorts of rules in the world of software, or of objects, where we're not supposed to access the data directly?

What about rules regarding access prileges? Shouldn't these be "protected" or "private"?

Answers: Python is very pragmatic. There is no "private," and there is no "protected." Everything is public and open. In such a case, why force people to go through a method to retrieve or set?

A good reason: Filter out bad values, either in getting or setting. But other than this, it's rare to have such methods in Python.

```python
class Person:
    def __init__(self, first_name, last_name, shoe_size):
        self.first_name = first_name  
        self.last_name = last_name    
        self.shoe_size = shoe_size    

    def fullname(self):
        return f'{self.first_name} {self.last_name}'
```

```python
p = Person('Reuven', 'Lerner', 46)
p.fullname()
```

# Why `self.name` and not just `name`?

When we say `self.name` in a method, what happens?

- Python looks for a local variable called `self`. It finds it -- the parameter, which is a local variable
- It retrieves the value that `self` is referring to -- in this case, the instance of `Company`
- It asks that instance of `Company` whether it has an attribute named `name`.
- It does! We get that value, and put it into the f-string.

```python
class Company:
    def __init__(self, name, domain, headcount):
        self.name = name
        self.domain = domain
        self.headcount = headcount
    def description(self):
        return f'Company {name}, domain is {self.domain}, with {self.headcount} people'
```

```python
python_co = Company('Python', 'python.org', 1000)
tiny_co = Company('Tiny', 'tiny.com', 1)
```

```python
print(python_co.description())
```

What happened here? Why did this fail?

- Python looks for a local variable named `name`. It doesn't find such a local variable.
- It then looks for a global variable `name`. It doesn't find one.
- It then asks if there's a `name` variable in the `builtins` scope. There isn't.
- We get an exception, saying that `name` is not defined.
