# What are objects?

Software is hard to write. And it's even harder to maintain.

In the 1970s, people already found this to be the case, and they started to look for ways to make software easier to write, and (especially) easier to maintain.

Alan Kay, at Xerox PARC, invented a programming language called Smalltalk. It was the first language to use "objects," which were a new paradigm meant to make it easier to write and maintain software.

Software was seen as very complex. Kay and his colleagues thought about what complex things existed in the world that they could perhaps make software more similar to. They decided to look at biological systems:

In biology, we have cells. Each cell has a particular type. Each type of cell knows how to send and receive different types of messages. If you know what kind of cell you're looking at, you know what its interface is to the world.

What if:

- We design our code using "objects," which are basically cells.
- Each object has as type, or a "class," which defines its behavior.
- Instead of sending/receiving messages, we'll talk about "invoking methods."

This was the paradigm shift that Kay promoted with Smalltalk.

A very large number of programming languages today now use this paradigm of objects, and they all owe a debt of gratitude to Smalltalk, even if they don't say so.

Python is one such language -- and some of the ideas and terminology in Python are actually taken from Smalltalk.

# Who cares? Or: So what?

We say that everything is an object in Python. What does this mean?

It means several things:

- The language is consistent. Everything follows the same rules, whether it's built-in stuff that comes with the language, or things that we add to the language as extensions.
- We can extend the language by writing new classes (i.e., new data types) that follow these rules

Once you understand how Python objects work, you'll see so many places in the core language that make more sense to you. 

Why are objects hard for people to learn?

- The vocabulary, which is different from the rest of programming
- It's treated as a religious discipline. It is a packaging, or logistical, technique, to make it easier to think about our software as we write, read, and maintain it.

# Object jargon/vocabulary

- `class` and `type`. Every value in Python is an object, and every object has a *type*. This means that every object's behavior is defined by certain rules, rules that are defined in the type. If I have a string, I know what a string does, because of the `str` *class*. In verbal Python, you can talk about "class" and "type" interchangeably. In Python code, they each have a specific role; you'll define new 
data types with the `class` keyword.
- You can think of a class as a factory for creating new objects, or new values. `str` is a factory for string values, and `dict` is a factory for dictionary values.
- When we talk about one particular string, we're talking about an *instance* of `str`. The class (type) is `str`, and the instance is one particular string value. In other words, the class (factory) produces many instances, each belonging to that class.
- Python has millions of objects running around at any given time. Every single one of them is an instance of some class. The class defines the behavior.

If I want to create a new value of type `X`, where `X` is the class/type, then I invoke `X()`. This gives me a new value of `X` back, aka a new instance of `X`.
    - I can create a new integer with `int()`
    - I can create a new string with `str()`
    - Many people say that `int` converts a string to an integer. But no! It creates a new instance of `int`, based on its argument (which can be a string).

- `instance` -- the word means, "a value whose class is ___," where ___ is a class.
- `object` -- this word is used in *far* too many different contexts. Every value in Python is an object. Every class is descended (through inheritance, which we'll discuss), from the top of the hierarchy, which is known as `object`. 

# What does the class matter?

If you know the class/type of a value, you know what it can do.

- If `x` is referring to a string, you can use `[]`, and `str.strip`, and `str.lower` on it.
- If `x` is referring to a `dict`, you can use `[]`, and `dict.get`, and `dict.pop` on it

Knowing the type is important. We'll be creating our own types, and can thus determine what people can do with our values.

# How can we reshape things into a class?

A class gives us a new data type. We can have whatever fields ("attributes") we want in each instance of our new type. We can call it whatever we want, too.

```python
class Person:    # we create a new type of data, known as Person, with the "class" keyword
    pass         # pass means -- I have nothing to say, but the indentation rules require something here
```

Syntax here:

- We use the keyword `class`
- We give the name of the class we're defining, which is normally capitalized or CamelCase
- It's OK to capitalize the name of a class
- Why, then, are `str` and `int` and `dict` not capitalized? They were created before those rules/conventions came into effect.

```python
# how do I create a new instance of Person? I invoke it!

Person()
```

```python
Person()
```

```python
Person()
```

I just got three different instances of `Person` back. It's like having three different lists, or strings -- they are all of the same type, but they are distinct objects.

Notice that the printed reprsentation ("repr") says that it's an instance of `Person` (defined in the `__main__` namespace), and then a bunch of hex digits -- the address in memory where the object is.

# Let's spice up our class

```python
class Person:
    def __init__(self, first_name):
        self.first_name = first_name

        
```

The above `Person` class is the first time we're seeing how to create a new class that does something semi-normal.

- Inside of our class definition, we used `def` to define a function
- However, any function defined inside of a class is known as a "method"
- Every method takes at least one argument, assigned to the first parameter, traditionally called `self`
- The job of the `__init__` method is to assign attributes to the newly created object in `self`. Here, we're assigning the value of the `first_name` variable to the `first_name` attribute on `self`
- Having done its job, `__init__` returns

# More slang!

If I say "dunder init," that's Python slang for `__init__` -- double underscore before and after the word "init". Dunder methods (like `__init__` are crucial in the Python ecosystem. If you define a dunder method, then either Python will run it automatically, or you're changing the default behavior. You are typically not going to run these methods yourself, directly.

```python
# let's define a more complex Person class

class Person:
    def __init__(self, first_name, last_name, shoe_size):
        self.first_name = first_name
        self.last_name = last_name
        self.shoe_size = shoe_size
```

```python
p = Person('Reuven', 'Lerner', 46)
```

```python
p.shoe_size
```

```python
p.last_name
```

```python
p.first_name
```

# Things are already better, even if we don't understand them yet

- This is a new `Person` class, with the same info about people as we had in the tuples
- Each instance of `Person` is a new `Person` object, with their own first and last names, as well as shoe dize
- If `p` is an instance of `Person`, then we can also say that `type(p)` will be `Person`
- We can retrieve any attribute set on `p` by `__init__` just by asking, e.g., `p.shoe_size` returns the size.

```python
# here, I'm defining the Person class -- I'm teaching Python how to create a new data structure
class Person:
    def __init__(self, name):
        self.name = name

# here, I'm invoking Person, the class
# this gives me back a new instance of Person
# I assign that instance to p

p = Person('Reuven')        
```

# Walk-through of `Person`

- We use `class` to tell Python we're defining a new type of data structure. We're teaching Python what a `class` is.
- We name our class with "CamelCase"
- We have a colon at the end of the line
- After the colon, starting on the next line, we have an indented block. This block is known as the "class body." This is where we tell Python what makes our class special and different from other classes in the language.
- Inside of the class body, we can define functions with `def`. But any function defined inside of the class is a *method*.
- The only method we defined here is `__init__` ("dunder init") We won't normally invoke this on our own, but we'll let Python invoke it for us.
- As it happens, this `__init__` method is invoked automatically right after the new `Person` object has been created. The new object gets created, Python automatically invokes `__init__`, and when that's done, it is returned to `p`.
- This new instance of `Person` can now exist in our program, and be used -- including retrieving the `name` attribute that we assigned in `__init__`.

# Attributes

What the heck is an attribute?

In Python, we know that we can store values in a *variable*. But it turns out that Python has *two* storage mechanisms. Variables are just one. The other is *attributes*.

Every single object in Python has attributes. They are a private dictionary on every object value. It doesn't look like a dict, but every attribute has a name and a value.

The names are normally written after a variable name, separated by a `.`. The values can be anything at all, no restrictions.

If I say

    a.b

this means, to Python, that I want to go to `a`, grab the value associated with that variable, and then turn the value and ask, "Do you have an attribute `b`?" If it does, then I want the value.

You can always tell an attribute, beacuse it comes after a `.`.
