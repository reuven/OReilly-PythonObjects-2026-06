# What are objects?

Software is hard to write. And it's even harder to maintain.

In the 1970s, people already found this to be the case, and they started to look for ways to make software easier to write, and (especially) easier to maintain.

Alan Kay, at Xerox PARC, invented a programming language called Smalltalk. It was the first language to use "objects," which were a new paradigm meant to make it easier to write and maintain software.

Software was seen as very complex. Kay and his colleagues thought about what complex things existed in the world that they could perhaps make software more similar to. They decided to look at biological systems:

In biology, we have cells. Each cell has a particular type. Each type of cell knows how to send and receive different types of messages. If you know what kind of cell you're looking at, you know what its interface is to the world.

What if:

- We design our code using "objects," which are basically cells.
- Each object has as type, or a "class," which defines its behavior.
- Instead of sending/receiving messages, we'll talk about "invoking methods."

This was the paradigm shift that Kay promoted with Smalltalk.

A very large number of programming languages today now use this paradigm of objects, and they all owe a debt of gratitude to Smalltalk, even if they don't say so.

Python is one such language -- and some of the ideas and terminology in Python are actually taken from Smalltalk.
